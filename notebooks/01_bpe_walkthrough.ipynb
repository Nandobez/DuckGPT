```python
# inspect a few learned merges as readable bytes
for new_id, (a, b) in zip(range(256, 256 + 8), merges[:8]):
    print(f"id {new_id}: {vocab[a]!r} + {vocab[b]!r} -> {vocab[new_id]!r}")
```

```python
sample = 'olá amigos do mundo, hello friends'
ids = encode(sample)
print(f'encoded ({len(ids)} ids):', ids)
print('decoded:', decode(ids))
```

## Compression

How much shorter is the BPE sequence than the raw byte sequence?

```python
raw = list(sample.encode('utf-8'))
print(f'raw bytes : {len(raw):>3}')
print(f'bpe tokens: {len(ids):>3}')
print(f'ratio     : {len(raw)/len(ids):.2f}x')
```
